```python
sales= float(input("enter sales value ($):"))
tax_rate = float(input("enter tax rate (%):"))

cogs= .4*sales
gross_profit= sales - cogs
taxes= gross_profit * tax_rate/100
net = gross_profit - taxes

print(f"sales: {sales}$")
print(f"cost of goods sold: {cogs}$")
print(f"gross profit: {gross_profit}$")
print(f"taxes: {taxes}$")
print(f"net income: {net}$")
```

```python
number1 =float(input("enter number:"))
number2 = float(input("enter number:"))
operator = (input("enter operator(eg. +,- ..):"))

if operator == "+":
    print(number1 + number2)
elif operator == "-":
    print(number1 - number2)
elif operator == "*":
    print(number1 * number2)
elif operator == "/":
    print(number1 / number2)
elif operator == "**":
    print(number1 ** number2)
elif operator == "//":
    print(number1 // number2)
elif operator == "%":
    print(number1 % number2)
else:
  print("invalid operator")
```

```python
print("hour fee")
for hour in range(1,9):
    fee = 5 + 2.5 * hour
    if fee < 10.0:
        fee = 10.0
    elif fee > 20.0:
        fee = 20.0
    print(f"{hour}    {fee}")
```

```python
ticketprice = 10
defaultAudience = 20
print("Ad$ Profit")
for adMoney in range(0,201,25):
  revenue = (defaultAudience  + 2* round( adMoney ** 0.5))* ticketprice

  cost = 200 + adMoney
  profit = revenue - cost
  print( f"{adMoney}   {profit}")
```
